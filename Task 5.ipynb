```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

fig, axes = plt.subplots(2, 2,  figsize=(12, 9))
fig.suptitle('Smoking Tendencies Analysis', fontsize=20, fontweight='bold', y=0.98)

# 1. Smoking by Gender (top-left)
gender_smoker = pd.crosstab(df['sex'], df['smoker'])
gender_smoker.plot(kind='bar', ax=axes[0, 0], color=['#3b82f6', '#ec4899'])
axes[0, 0].set_title('Smokers by Gender', fontsize=10, fontweight='bold')
axes[0, 0].set_xlabel('Gender', fontsize=12)
axes[0, 0].set_ylabel('Count', fontsize=12)
axes[0, 0].grid(axis='y', linestyle='--', alpha=0.7)
axes[0, 0].legend(['Non-smoker', 'Smoker'])





age_smoker = pd.crosstab(df['age'], df['smoker'])
age_smoker.plot(kind='bar', ax=axes[0, 1], color=['#3b82f6', '#ec4899'])
axes[0, 1].set_title('Smokers by Age Group', fontsize=14, fontweight='bold')
axes[0, 1].set_xlabel('Age', fontsize=12)
axes[0, 1].set_ylabel('Count', fontsize=12)
axes[0, 1].grid(axis='y', linestyle='--', alpha=0.7)
axes[0, 1].legend(['Non-smoker', 'Smoker'])

region_smoker = pd.crosstab(df['region'], df['smoker'])
region_smoker.plot(kind='bar', ax=axes[1, 0], color=['#3b82f6', '#ec4899'])
axes[1, 0].set_title('Smokers by Region', fontsize=10, fontweight='bold')
axes[1, 0].set_xlabel('Region', fontsize=12)
axes[1, 0].set_ylabel('Count', fontsize=12)
axes[1, 0].grid(axis='y', linestyle='--', alpha=0.7)
axes[1, 0].legend(['Non-smoker', 'Smoker'])
```

### Key Observations


<br>1. Male smokers > Female Smokers
<br>2. Southwest Region has the highest Smokers compared to rest.
<br>3. Smoking age begins from min 21 and majority smokers have age of 19.
